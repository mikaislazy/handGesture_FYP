```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Generate fake data
num_iterations = 10
error_rate = np.random.uniform(0, 100, num_iterations)
score = np.random.uniform(0, 5, num_iterations)
time_intervals = np.random.uniform(5, 15, num_iterations)


# Generate x-axis values
iterations = np.arange(1, num_iterations + 1)

# Create line chart
plt.figure(figsize=(6, 3))

# Plot error rate
plt.plot(iterations, error_rate, label='Error Rate', marker='o')

# Set labels and title
plt.xlabel('Trial')
plt.ylabel('Percentage')
plt.title('Error Rate Over Time')


# Show legend
plt.legend()

# Show the plot
plt.show()
#########################
# Create line chart
plt.figure(figsize=(6, 3))

# Plot success rate
plt.plot(iterations, score, label='Score', marker='o')
# Set labels and title
plt.xlabel('Trial')
plt.ylabel('Score')
plt.title('Score in Task1')

# Show legend
plt.legend()

# Show the plot
plt.show()



#########################
# Create line chart
plt.figure(figsize=(6, 3))
# Plot normalized time intervals
plt.plot(iterations,time_intervals, label='Time Interval', marker='o')

# Set labels and title
plt.xlabel('Trial')
plt.ylabel('Time Interval')
plt.title('Time Intervals in Task2')

# Show legend
plt.legend()

# Show the plot
plt.show()


```
